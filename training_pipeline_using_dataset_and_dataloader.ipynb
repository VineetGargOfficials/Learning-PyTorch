```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-cancer-Detection/refs/heads/master/data.csv")
df.head()
```

```python
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Y_train = encoder.fit_transform(Y_train)
Y_test = encoder.transform(Y_test)
```

```python
import torch
```

```python
X_train = torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
Y_train = torch.from_numpy(Y_train).float()
Y_test = torch.from_numpy(Y_test).float()
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)
    
    def __getitem__(self, index):
        return self.features[index], self.labels[index]
        
```

```python
train_dataset = CustomDataset(X_train, Y_train)
test_dataset = CustomDataset(X_test, Y_test)
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
import torch.nn as nn
```

```python
class MySimpleNN(nn.Module):

    def __init__(self, num_features):
        super().__init__()
        self.linear = nn.Linear(num_features, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear(features)
        out = self.sigmoid(out)

        return out
    
loss_function = nn.BCELoss()
```

```python
learning_rate = 0.1
epoch = 25
```

```python
model = MySimpleNN(X_train.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr= learning_rate)
```

```python
for epoch in range(epoch):
    for batch_features, batch_labels in train_loader:
        
        y_pred = model(batch_features)

        loss = loss_function(y_pred, batch_labels.view(-1,1))

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

    print(f"Epoch: {epoch + 1}, Loss: {loss.item()}")
```

```python
model.eval()
accuracy_list = []
with torch.no_grad():
    for batch_features, batch_labels in test_loader:
        # Forward pass
        y_pred = model(batch_features)              # shape: (N, 3)
        predicted_classes = (y_pred > 0.5).float().squeeze()  # âœ… Threshold for binary classification  # shape: (N,)
        
        # Compare predictions with labels
        batch_accuracy = (predicted_classes == batch_labels).float().mean().item()
        accuracy_list.append(batch_accuracy)

overall_accuracy = sum(accuracy_list) / len(accuracy_list)

print(f"Test Accuracy: {overall_accuracy*100:.2f}%")
```
