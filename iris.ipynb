```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.drop(columns="Id", inplace=True)
```

```python
df.head()
```

```python
X_train, Y_train, X_test, Y_test = train_test_split(df.iloc[:, :4], df.iloc[:, 4], test_size=0.2)
```

```python
encoder = LabelEncoder()
Y_test = encoder.fit_transform(Y_test)
X_test = encoder.fit_transform(X_test)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
Y_train = scalar.fit_transform(Y_train)
```

```python
import torch
```

```python
X_train_tensor = torch.from_numpy(X_train)
Y_train_tensor = torch.from_numpy(Y_train)

Y_test_tensor = torch.from_numpy(Y_test)
X_test_tensor = torch.from_numpy(X_test)
```

```python
X_train_tensor
```

```python
class MySimpleNN():
    def __init__(self, X):
        self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
        self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

    def forward_bias(self, X):
        z = torch.matmul(X, self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred
    
    def loss_function(self, y_pred, y_true):
        eps = 1e-7
        y_pred = torch.clamp(y_pred, eps, 1 - eps)
        loss = -(y_true * torch.log(y_pred) + (1 - y_true)* torch.log(1 - y_pred)).mean()
        return loss
    
```

```python
learning_rate = 0.1
epochs = 25

model = MySimpleNN(X_train_tensor)

for epochs in range(epochs):
    y_pred = model.forward_bias(X_train_tensor)
    loss = model.loss_function(y_pred, X_test_tensor)
    # print(f'Epoch: {epochs}, Loss: {loss.item()}')

    loss.backward()

        # parameter update
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad
    
    # Zero gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()

    # Print loss in each epoch
    
    print(f'Epoch: {epochs + 1}, Loss: {loss.item()}')


    
    
```
