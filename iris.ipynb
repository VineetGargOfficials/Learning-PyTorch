```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('Iris.csv')
df.head()
```

```python
df.drop(columns="Id", inplace=True)
```

```python
df.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:, :4], df.iloc[:, 4], test_size=0.2)
```

```python
encoder = LabelEncoder()
Y_test = encoder.fit_transform(Y_test)
Y_train = encoder.fit_transform(Y_train)
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.fit_transform(X_test)
```

```python
import torch
```

```python
X_train_tensor = torch.from_numpy(X_train).float()
Y_train_tensor = torch.from_numpy(Y_train).long()
X_test_tensor = torch.from_numpy(X_test).float()
Y_test_tensor = torch.from_numpy(Y_test).long()
```

```python
X_train_tensor
```

```python
import torch.nn as nn

class MySimpleNN(nn.Module):
    def __init__(self, num_features, num_outputs=3):

        super().__init__()
        self.linear = nn.Linear(num_features, num_outputs)
        # self.sigmoid = nn.Sigmoid()

    def forward(self, features):
        out = self.linear(features)
        # out = self.sigmoid(out)

        return out
        
loss_function = nn.CrossEntropyLoss()
```

```python
learning_rate = 0.1
epochs = 25

model = MySimpleNN(X_train_tensor.shape[1], num_outputs=3)

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epochs in range(epochs):
    y_pred = model(X_train_tensor)
    loss = loss_function(y_pred, Y_train_tensor)
    # print(f'Epoch: {epochs}, Loss: {loss.item()}')
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


    # Print loss in each epoch
    
    print(f'Epoch: {epochs + 1}, Loss: {loss.item()}')


    
    
```

```python
model.eval()
with torch.no_grad():
    logits = model(X_test_tensor)           # shape: (N, 3)
    y_pred = torch.argmax(logits, dim=1)    # predicted class indices, shape: (N,)

# Ensure targets are long and 1-D
y_true = Y_test_tensor.view(-1).long()
correct = (y_pred == y_true).sum().item()
accuracy = correct / y_true.size(0)

print(f"Test Accuracy: {accuracy*100:.2f}%")
```
